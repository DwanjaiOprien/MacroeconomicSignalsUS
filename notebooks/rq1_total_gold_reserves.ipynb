<h3><center>Data Preparation and Loading</h3></center>

```python
# Data packages 
from pathlib import Path 
import pandas as pd #For data processing and cleaning 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
DATA_DIR = Path("..")/"datasets"
totalReserves = pd.read_csv(DATA_DIR/ "TRESEGCNM052N.csv")
```

```python
from pathlib import Path
Path(".").resolve()
```

```python
# Indexing using Pandas date_time function 
# totalReserves = totalReserves.set_index(totalReserves['observation_date']) --> Just makes another observation column 
# df['year'] = df['datetime_obj'].dt.year
totalReserves['observation_date'] = pd.to_datetime(
    totalReserves['observation_date']
)

totalReserves['year'] = totalReserves['observation_date'].dt.year

print(totalReserves.head())
```

<h3><center><Preparing the Model </h3></center>

<h3><center>Preparing the Model </h3></center>

```python
plt.hist(totalReserves['TRESEGCNM052N'])
plt.title("Total Gold Reserves (2015-2025)")
plt.xlabel("Year")
plt.ylabel("Amount")
plt.grid(True)
plt.show()

# A histogram is not recommended for a plot of this scale. 
```

<h3><center>Data Visualisation </h3></center>

```python
yearly_gold = (
    totalReserves
        .groupby('year')['TRESEGCNM052N']
        .mean()
        .reset_index()
)

yearly_gold['gold_reserves_millions'] = yearly_gold['TRESEGCNM052N'] / 1_000_000


# rescale
yearly_gold['gold_reserves_millions'] = yearly_gold['TRESEGCNM052N'] / 1_000_000

plt.figure(figsize=(10, 5))
plt.plot(
    yearly_gold['year'],
    yearly_gold['gold_reserves_millions'],
    marker='o'
)
plt.xlabel("Year")
plt.ylabel("Total Gold Reserves (Millions)")
plt.title("Total U.S. Gold Reserves Over Time (2015â€“2025)")
plt.grid(True)
plt.savefig("gold_reserves_2015_2025.png", dpi=300, bbox_inches="tight")
plt.show()


```

                                ## Data Interpretation 
Total U.S. gold reserves increased between 2020 and 2022, following a relatively stable period prior to 2020. This increase coincides temporally with the COVID-19 pandemic period. While the timing of this overlap is notable, the analysis remains descriptive and does not establish a causal relationship between the pandemic and changes in gold reserves. Gold reserves continue to rise in the post-COVID period, extending through 2025. Earlier in the series, a noticeable decline occurs between 2015 and 2016, followed by a period of relative stabilization during 2017â€“2018.Further analysis incorporating policy actions or complementary macroeconomic indicators would be required to assess potential explanatory mechanisms.
